# Getting Financial Data

```python
#attempting to get 5 years stock data for tesla, amazon, apple, microsoft and google
import requests
api_key='f59b1c7bdca2d442c8a68a4169f6faa4' #use own api key, maybe we can create one with the bt4222 email
start_date = '2017-01-01'
end_date = '2022-02-09'
stockName = ['TSLA','AMZN','AAPL','MSFT','GOOG']

financial_data_csv = {}

marketcap = {}


for comp in stockName:
  financial_data_csv[comp] = requests.get(f"https://financialmodelingprep.com/api/v3/income-statement/{comp}?period=annual&limit=400&apikey={api_key}").json()

for comp in stockName:
  marketcap[comp] = requests.get(f"https://financialmodelingprep.com/api/v3/key-metrics/{comp}?limit=6&apikey={api_key}").json()


```

```python
import pandas as pd

df = pd.DataFrame()
for comp_name in financial_data_csv.keys():
  try:
    fin_data = financial_data_csv[comp_name]
    tempdf = pd.DataFrame.from_dict(fin_data)
    tempdf['Stock Name'] = comp_name
    df = pd.concat([df,tempdf])
  except:
    continue
```

```python
import pandas as pd

df_marketcap = pd.DataFrame()
for comp_name in marketcap.keys():
  try:
    fin_data = marketcap[comp_name]
    tempdf = pd.DataFrame.from_dict(fin_data)
    tempdf['Stock Name'] = comp_name
    df_marketcap = pd.concat([df_marketcap,tempdf])
  except:
    continue

# financial_data_csv.keys()
```
